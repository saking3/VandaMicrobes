```python
import pandas as pd
from scipy.stats import ttest_ind

# Load your Excel file
df = pd.read_excel('G:/Official_Vanda_Organizing/Spatial_Morphologic_Data/1_Vanda_Environmental_Data/t_test_data.xlsx')

# Clean column names: remove spaces, punctuation, etc.
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Convert 'Webbed' to a boolean: True for 'Y', False for 'N'
df['Webbed'] = df['Webbed'].map({'Y': True, 'N': False})

# Define the two groups
webbed = df[df['Webbed']]
nonwebbed = df[~df['Webbed']]

# Variables to test
variables = ['Depth', 'Irradiance', 'Shoreline', 'Sedimentation', 'Conductivity']

# Run t-tests and store results
print("T-Test Results (Webbed vs. Non-Webbed):\n")
for var in variables:
    t_stat, p_val = ttest_ind(webbed[var], nonwebbed[var], equal_var=False)
    direction = "higher" if webbed[var].mean() > nonwebbed[var].mean() else "lower"
    print(f"{var}: t-stat = {t_stat:.3f}, p-value = {p_val:.3f}, webbed mean is {direction}")

# Boxplots
df['Webbed_Label'] = df['Webbed'].map({True: 'Webbed', False: 'Nonwebbed'})
for var in variables:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Webbed_Label', y=var, data=df, palette='Set2')
    plt.title(f"{var} by Webbed Status")
    plt.tight_layout()
    plt.show()
```
